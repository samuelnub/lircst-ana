```python
#@title Let's use this notebook to compress the sinogram data

import numpy as np
import matplotlib.pyplot as plt
import os

if False: # Dangerous to run this cell, it will overwrite the sinogram files
    # Get all directories within ../data folder
    data_dir = '../data'
    dirs = [d for d in os.listdir(data_dir) if os.path.isdir(os.path.join(data_dir, d))]
    print(f"Our phantom directories are {dirs}")

    # Get all files within a directory
    def get_files_in_dir(dir_name):
        return [f for f in os.listdir(os.path.join(data_dir, dir_name)) if os.path.isfile(os.path.join(data_dir, dir_name, f))]

    for dir_name in dirs:
        files = get_files_in_dir(dir_name)
        sino_files = [f for f in files if f.startswith('sino')]
        # Open each sinogram file and sum its last axis
        for sino_file in sino_files:
            file_path = os.path.join(data_dir, dir_name, sino_file)
            print(f"Processing {file_path}")
            sino = np.load(file_path)
            if sino.shape[-1] > 1:
                # Sum along the last axis
                compressed_sino = np.sum(sino, axis=-1)
                # Save the compressed sinogram
                compressed_file_path = os.path.join(data_dir, dir_name, f'{sino_file}')
                with open('{}'.format(compressed_file_path), 'wb') as f:
                    np.save(compressed_file_path, compressed_sino)
                print(f"Saved compressed sinogram to {compressed_file_path}")
            else:
                print(f"Skipping {sino_file} as it does not have the expected shape.")
        print(f"Processed {len(sino_files)} sinogram files in {dir_name} directory.")
```
